## Observations and Insights 

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
combined_df = pd.merge(mouse_metadata, study_results, how='outer', on='Mouse ID')
# Display the data table for preview
combined_df
```

```python
# Checking the number of mice.
combined_df.count()
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary dataframe.

mean = combined_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].mean()
median = combined_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].median()
variance = combined_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].var()
stdv = combined_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].std()
sem = combined_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].sem()
statistics_df = pd.DataFrame({"Mean": mean, "Median": median, "Variance": variance, "Standard Deviation": stdv, 
                          "SEM": sem})
statistics_df
```

## Bar and Pie Charts

```python
# Generate a bar plot showing the total number of timepoints for all mice tested for each drug regimen using Pandas.
group1_df = pd.DataFrame(combined_df.groupby(["Drug Regimen"]).count()).reset_index()

#Set the dataframe to two columns
regimen_df = group1_df[["Drug Regimen", 'Mouse ID']]

#Set the index so the drug regimen labels will appear
regimen_df = regimen_df.set_index("Drug Regimen")

#Barchart
regimen_df.plot(kind = "bar", figsize =(10,3))

plt.title("Drug Regimen Count")
plt.show()
plt.tight_layout()
```

```python
# Generate a bar plot showing the total number of timepoints for all mice tested for each drug regimen using pyplot.

#get the data for the total number of timepoints and list of drugs
timepoints_df = (combined_df.groupby(["Drug Regimen"])["Age_months"].count()).tolist()
drugs_df = statistics_df.index.tolist()

x_axis = np.arange(len(timepoints_df))

# Set X-Axis
x_axis = timepoints_df

#Barchart
plt.figure(figsize=(10,3))
plt.bar(x_axis, timepoints_df, color='b', alpha=.5, align="center")

#Set labels 
plt.title("Drug Regimen Count")
plt.xlabel("Drug Regimen")
plt.ylabel("Count")
```

```python
# Generate a pie plot showing the distribution of female versus male mice using Pandas

#Group data by gender
gender_df = pd.DataFrame(combined_df.groupby(["Sex"]).count()).reset_index()

#Set the dataframe to two columns
gender_df = gender_df[["Sex", "Mouse ID"]]

#Pie Plot
plt.figure(figsize=(12,6))
ax1 = plt.subplot(121, aspect="equal")
gender_df.plot(kind="pie", y = "Mouse ID", ax=ax1, autopct='%1.1f%%',
              startangle=190, shadow=True, labels=gender_df["Sex"], legend = False, fontsize=14)

#Set labels
plt.title("Gender Breakdown")
plt.xlabel("")
plt.ylabel("")
```

```python
plt.clf()
plt.cla()
plt.close()
```

## Line and Scatter Plots

```python
# Generate a line plot of tumor volume vs. time point for a mouse treated with Capomulin
mouse_data = combined_df[combined_df["Mouse ID"].isin(["m601"])]
mouse_data

time_vs_tumer = mouse_data[["Mouse ID", "Timepoint", "Tumor Volume (mm3)"]]
time_vs_tumer

plot_df = time_vs_tumer.reset_index()
plot_df

line_plot = plot_df[["Mouse ID", "Timepoint", "Tumor Volume (mm3)"]]
line_plot

lines = line_plot.plot.line()
```

```python
# Generate a scatter plot of average tumor volume vs. mouse weight for the Capomulin regimen
capomulin = combined_df[combined_df["Drug Regimen"].isin(["Capomulin"])]

average = pd.DataFrame(capomulin.groupby(["Mouse ID", "Weight (g)"])["Tumor Volume (mm3)"].mean()).reset_index()

#Set the index
average =average.reset_index()

#Scatter Plot
average.plot(kind="scatter", x="Weight (g)", y="Tumor Volume (mm3)")
plt.show()
```
